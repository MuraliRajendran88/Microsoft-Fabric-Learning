```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
file_name="fromgithub"
file_path="Files/DataFiles/fromgithub"
df=spark.read.csv(file_path,header=True,inferSchema=True)
df.show()
display(df)
```

```python
##add new column with dataframe
import sys
from pyspark.sql.functions import col,when
df_transformed=df.withColumn(
    "subject1_status",
when(col("subject_1")>35,"pass").otherwise("fail")
).withColumn(
    "subject2_status",
    when(col("subject_2")>35,"pass").otherwise("fail")
).withColumn(
    "subject3_status",
    when(col("subject_3")>35,"pass").otherwise("fail")
)
df_transformed.show()
display(df_transformed)

   
```

```python
from pyspark.sql import functions as f
max_value=df_transformed.agg(f.max("subject_2")).show()
display(max_value)

min_value=df_transformed.agg(f.min("subject_3")).show()
display(min_value)
 
```

```python
from delta.tables import DeltaTable

# Write the transformed data into the Silver layer
df_transformed.write.option("overwriteSchema", "true").format("delta").mode("overwrite").save("Tables/student_data")
     
```
